```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
a = np.random.random((10, 10))
```

```python
b = np.random.random((10,10))
```

```python
def f(x, y):
    return x**2*y
```

```python
z = f(a, b)
```

```python
gy, gx = np.gradient(z)
```

```python
plt.pcolor(a, b, z)
plt.colorbar()
plt.quiver(a, b, gx, gy)
plt.show()
```
